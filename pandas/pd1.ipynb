```python
# 数据当中，有g为单位，也有mg为单位的，查找g为单位的列
col_names = food_info.columns.tolist()
print (col_names)
gram_columns = []

for c in col_names:
    if c.endswith("(g)"):
        gram_columns.append(c)
gram_df = food_info[gram_columns]
print (gram_df)
```
